```python
### Preprocessing For classification
import pandas as pd
mdf = pd.read_csv('../Common/rly_final_movies.tsv', sep='\t')
emdf = pd.read_csv('../Common/filtered_final_movies_5.tsv', sep='\t')
bmdf = pd.read_csv('../Common/moviesWithBinaryValues.tsv', sep='\t')
pdf = pd.read_csv('../topic_discovery/lda_topics.tsv', sep='\t')


emdf = emdf.drop(columns = ["primaryTitle","startYear","runtimeMinutes","overview","original_language","release_date","keywords","synopsis","domestic","international","worldwide","distributor","domestic_opening","mpaa","genres","budget","actors","writers","directors"])

mdf = mdf.merge(bmdf,how='left',on='tconst')
mdf = mdf.merge(pdf,how='left',on='tconst')
mdf = mdf.merge(emdf,how='left',on='tconst')
##Transmedia
mdf['Transmedia'] = mdf['Transmedia'].map({'YES': 1, 'NO': 0, 'UNKNOWN': 0})

##Adaptation
mdf['Adaptation'] = mdf['Adaptation'].map({'YES': 1, 'NO': 0, 'UNKNOWN': 0})

## Keep Sequels
mdf['Sequel'] = mdf['Sequel'].map({'YES': 1, 'NO': 0, 'UNKNOWN': 0})



# ##Remove sequels
# mdf = mdf.merge(sequel[['tconst','Response']],how='left',on='tconst')
# mdf = mdf[~(mdf["Response"] == "YES")]
# mdf = mdf.drop(columns = ['Response'])


## Drop unused features
mdf = mdf.drop(
    ["primaryTitle" ,"overview", "tconst", "release_date","keywords","synopsis","genres","actors", "genres_vector", "directors", "writers", "distributor","profit"], axis=1
)

## Only keep movies with mpaa ratings
# mdf = mdf[mdf["mpaa"].isin(["G","PG","PG-13", "R", "NC-17"])]
mdf["mpaa"] = mdf["mpaa"].replace({"Not Rated": "Unrated","TV-G":"G", "TV-PG":"PG", "TV-14":"PG-13", "TV-MA":"R", "TV-Y7":'G'})
mdf["mpaa"] = mdf["mpaa"].fillna("Unrated")
mdf = pd.get_dummies(mdf, columns=['mpaa'], prefix='mpaa')

# # Define the ordinal mapping
# mpaa_mapping = {'G': 0, 'PG': 1, 'PG-13': 2, 'R': 3, 'NC-17': 4}
# mdf['mpaa'] = mdf['mpaa'].map(mpaa_mapping)

mdf["day_of_week"] = mdf["day_of_week"].map({"Monday": 1, "Tuesday": 2, "Wednesday": 3, "Thursday": 4, "Friday": 5, "Saturday": 6, "Sunday": 7})
mdf.to_csv("./preprocessed.tsv", sep="\t", index=False)
```
